# Visualizing earnings based on college majors

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

## Import the data

```python
# Read the dataset
recent_grads = pd.read_csv("recent-grads.csv")
```

Description of the columns:
 * `Rank` - Rank by median earnings (the dataset is ordered by this column).
 * `Major_code` - Major code.
 * `Major` - Major description.
 * `Major_category` - Category of major.
 * `Total` - Total number of people with major.
 * `Sample_size` - Sample size (unweighted) of full-time.
 * `Men` - Male graduates.
 * `Women` - Female graduates.
 * `ShareWomen` - Women as share of total.
 * `Employed` - Number employed.
 * `Median` - Median salary of full-time, year-round workers.
 * `Low_wage_jobs` - Number in low-wage service jobs.
 * `Full_time` - Number employed 35 hours or more.
 * `Part_time` - Number employed less than 35 hours.

## Scatter plots

```python
fig = plt.figure(figsize=(15,20))

ax1 = fig.add_subplot(3,2,1)
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', ax=ax1, title='Median vs. Sample_size')

ax2 = fig.add_subplot(3,2,2)
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', ax=ax2, title='Unemployment_rate vs. Sample_size')

ax3 = fig.add_subplot(3,2,3)
recent_grads.plot(x='Full_time', y='Median', kind='scatter', ax=ax3, title='Median vs. Full_time')

ax4 = fig.add_subplot(3,2,4)
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', ax=ax4, title='Unemployment_rate vs. ShareWomen')

ax5 = fig.add_subplot(3,2,5)
recent_grads.plot(x='Men', y='Median', kind='scatter', ax=ax5, title='Median vs. Men')

ax6 = fig.add_subplot(3,2,6)
recent_grads.plot(x='Women', y='Median', kind='scatter', ax=ax6, title='Median vs. Women')
```

## Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(15,25))
i = 0

for col  in cols:

    i += 1

    ax = fig.add_subplot(4, 2, i)

    recent_grads[col].hist(ax=ax)
    ax.set_title(col);
```
