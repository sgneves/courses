```python
# 1. Recap
import pandas as pd
import matplotlib.pyplot as plt

unrate = pd.read_csv("unrate.csv")

unrate['DATE'] = pd.to_datetime(unrate['DATE'])

plt.plot(unrate.loc[:11,"DATE"], unrate.loc[:11,"VALUE"])

plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948");
```

```python
# 5. Adding Data
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.plot(unrate.loc[:11,"DATE"], unrate.loc[:11,"VALUE"])

ax2 = fig.add_subplot(2,1,2)
ax2.plot(unrate.loc[12:23,"DATE"], unrate.loc[12:23,"VALUE"]);
```

```python
# 6. Formatting And Spacing
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(2,1,1)
ax1.plot(unrate.loc[:11,"DATE"], unrate.loc[:11,"VALUE"])

ax2 = fig.add_subplot(2,1,2)
ax2.plot(unrate.loc[12:23,"DATE"], unrate.loc[12:23,"VALUE"]);
```

```python
# 7. Comparing Across More Year
fig = plt.figure(figsize=(12,12))

for i in range(5):
    
    ax = fig.add_subplot(5,1,i+1)

    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    
    ax.plot(subset['DATE'], subset['VALUE'])
```

```python
# 8. Overlaying Line Charts
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(6,3))

plt.plot(unrate.loc[:11,"MONTH"], unrate.loc[:11,"VALUE"], c="red")
plt.plot(unrate.loc[12:23,"MONTH"], unrate.loc[12:23,"VALUE"], c="blue");
```

```python
# 9. Adding More Lines
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):

    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i])
```

```python
# 10. Adding A Legend
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):

    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=str(1948+i))
    
plt.legend(loc='upper left');
```

```python
# 11. Final Tweaks
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):

    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=str(1948+i))
    
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.legend(loc='upper left');
plt.title("Monthly Unemployment Trends, 1948-1952");
```
