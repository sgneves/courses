```python
# 3. Introduction To The Data
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

plt.plot(women_degrees["Year"], women_degrees["Biology"]);
```

```python
# 4. Visualizing The Gender Gap
fig, ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], c="blue")
ax.plot(women_degrees["Year"], 100 - women_degrees["Biology"], c="green")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(['Women','Men'], loc='upper right');
```

```python
# 6. Hiding Tick Marks
fig, ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], c="blue")
ax.plot(women_degrees["Year"], 100 - women_degrees["Biology"], c="green")
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.legend(['Women','Men'], loc='upper right')
ax.set_title('Percentage of Biology Degrees Awarded By Gender');
```

```python
# 7. Hiding Spines
fig, ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], c="blue")
ax.plot(women_degrees["Year"], 100 - women_degrees["Biology"], c="green")
ax.tick_params(bottom=False, top=False, left=False, right=False)
for axis in ["bottom", "top", "left", "right"]:ax.spines[axis].set_visible(False)
ax.legend(['Women','Men'], loc='upper right')
ax.set_title('Percentage of Biology Degrees Awarded By Gender');
```

```python
# 8. Comparing Gender Gap Across Degree Categories
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

fig = plt.figure(figsize=(12, 12))
i = 0

for cat in major_cats:
    
    i += 1
    
    ax = fig.add_subplot(2, 2, i)
    
    ax.plot(women_degrees["Year"], women_degrees[cat], c="blue")
    ax.plot(women_degrees["Year"], 100 - women_degrees[cat], c="green")
    
    ax.axis([1968, 2011, 0, 100])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for axis in ["bottom", "top", "left", "right"]:ax.spines[axis].set_visible(False)
    if i==4: ax.legend(['Women','Men'], loc='upper right')
    ax.set_title(cat);
```
