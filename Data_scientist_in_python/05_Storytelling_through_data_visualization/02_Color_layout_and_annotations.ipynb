```python
# 1. Introduction
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

fig = plt.figure(figsize=(12, 12))
i = 0

for cat in major_cats:
    
    i += 1
    
    ax = fig.add_subplot(2, 2, i)
    
    ax.plot(women_degrees["Year"], women_degrees[cat], c="blue")
    ax.plot(women_degrees["Year"], 100 - women_degrees[cat], c="green")
    
    ax.axis([1968, 2011, 0, 100])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for axis in ["bottom", "top", "left", "right"]:ax.spines[axis].set_visible(False)
    ax.set_title(cat)

ax.legend(['Women','Men'], loc='upper right');
```

```python
# 3. Setting Line Color Using RGB
blue = (0/255,107/255,164/255)
orange = (255/255,128/255,14/255)

fig = plt.figure(figsize=(12, 12))
i = 0

for cat in major_cats:
    
    i += 1
    
    ax = fig.add_subplot(2, 2, i)
    
    ax.plot(women_degrees["Year"], women_degrees[cat], c=blue)
    ax.plot(women_degrees["Year"], 100 - women_degrees[cat], c=orange)
    
    ax.axis([1968, 2011, 0, 100])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for axis in ["bottom", "top", "left", "right"]:ax.spines[axis].set_visible(False)
    ax.set_title(cat)

ax.legend(['Women','Men'], loc='upper right');
```

```python
# 4. Setting Line Width
fig = plt.figure(figsize=(12, 12))
i = 0

for cat in major_cats:
    
    i += 1
    
    ax = fig.add_subplot(2, 2, i)
    
    ax.plot(women_degrees["Year"], women_degrees[cat], c=blue, linewidth=3)
    ax.plot(women_degrees["Year"], 100 - women_degrees[cat], c=orange, linewidth=3)
    
    ax.axis([1968, 2011, 0, 100])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for axis in ["bottom", "top", "left", "right"]:ax.spines[axis].set_visible(False)    
    ax.set_title(cat)

ax.legend(['Women','Men'], loc='upper right');
```

```python
# 5. Improve the Layout and Ordering
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    
    ax = fig.add_subplot(1, 6, sp+1)

    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=orange, label='Men', linewidth=3)

    for key,spine in ax.spines.items():
        spine.set_visible(False)

    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")

plt.legend(loc='upper right')
plt.show()
```

```python
# 7. Annotating in Matplotlib
fig = plt.figure(figsize=(18, 3))
axes = []
for sp in range(0,6):
    
    ax = fig.add_subplot(1, 6, sp+1)
    axes.append(ax)

    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=orange, label='Men', linewidth=3)

    for key,spine in ax.spines.items():
        spine.set_visible(False)

    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")

axes[0].text(2005, 87, "Men")
axes[0].text(2002, 8, "Women")
axes[-1].text(2005, 62, "Men")
axes[-1].text(2001, 35, "Women");
```
